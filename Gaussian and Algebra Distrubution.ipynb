```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Generate data: 1000 students with mean height=165 cm, std deviation=10 cm
heights = np.random.normal(loc=165, scale=10, size=1000)

# Plot the distribution
sns.histplot(heights, kde=True)
plt.title('Gaussian Distribution of Student Heights')
plt.xlabel('Height (cm)')
plt.ylabel('Frequency')
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

## Generate Gaussian data: mean = 70, std = 10, 1000 students
scores = np.random.normal(loc=70, scale=10, size=1000)

#Plot the distrubution
sns.histplot(scores,kde=True)
plt.title('Gaussian distrubution of exam scores')
plt.xlabel('Score')
plt.ylabel('Frequency')
plt.legend()
plt.show()
```

```python
#Step 2: Shifted Data
# Add 5 marks to every student
shifted_scores = scores + 5

sns.histplot(shifted_scores, kde=True, color='orange', label='Shifted')
plt.title('Shifted Distribution (+5 Marks)')
plt.xlabel('Score')
plt.ylabel('Frequency')
plt.legend()
plt.show()
```

```python
 #Scaled Data (20% Bonus)
 # Give 20% bonus to everyone
scaled_scores = scores * 1.2

sns.histplot(scaled_scores, kde=True, color='green', label='Scaled')
plt.title('Scaled Distribution (1.2)')
plt.xlabel('Score')
plt.ylabel('Frequency')
plt.legend()
plt.show()
```

```python
#Standardized Data (Z-score)
# Standardize: Z = (X - ) / 
standardized_scores = (scores - np.mean(scores)) / np.std(scores)

sns.histplot(standardized_scores, kde=True, color='purple', label='Standardized')
plt.title('Standardized Distribution (Z-Score)')
plt.xlabel('Z')
plt.ylabel('Frequency')
plt.legend()
plt.show()
```
